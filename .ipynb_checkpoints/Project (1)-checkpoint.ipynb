### 5. For each station (itcs_stopName), determine the average number of passengers

```python
passenger_data = defaultdict(lambda: {'total_passengers': 0, 'stop_count': 0})

for file in os.listdir(csv_folder): 
    if file.endswith('.csv'): 
        file_path = os.path.join(csv_folder, file) 
        
df = pd.read_csv(file_path, usecols=['itcs_stopName', 'itcs_numberOfPassengers'])

stop_group = df.groupby('itcs_stopName')['itcs_numberOfPassengers'].agg(['sum', 'count']) 

for stop, row in stop_group.iterrows(): 
    passenger_data[stop]['total_passengers'] += row['sum'] 
    passenger_data[stop]['stop_count'] += row['count'] 

average_passengers_per_stop = {stop: data['total_passengers'] / data['stop_count'] 
                               for stop, data in passenger_data.items()} 

average_passengers_df = pd.DataFrame.from_dict(average_passengers_per_stop, orient='index', columns=['average_passengers']) 
average_passengers_df.sort_index(inplace=True)

```

```python
average_passengers_df
```

### 6. For each station, determine the buses that have stopped there at least once.

```python
file_to_bus = dict(zip(df_in['name'], df_in['busNumber']))

stops_to_buses = defaultdict(set)

for file in os.listdir(csv_folder): 
    if file.endswith('.csv'): 
        bus_number = file_to_bus.get(file.split('.')[0], None) 
        if bus_number is not None: 
            file_path = os.path.join(csv_folder, file) 
            df = pd.read_csv(file_path, usecols=['itcs_stopName']) 
            for stop in df['itcs_stopName'].unique(): 
                stops_to_buses[stop].add(bus_number) 
                
stops_to_buses_df = pd.DataFrame([(stop, list(buses)) for stop, buses in stops_to_buses.items()], columns=['StopName', 'Buses'])   
```

```python
stops_to_buses_df
```

### 7. For each station, determine the buses that have stopped there at least ten times.

```python
stops_to_buses_counts = defaultdict(lambda: defaultdict(int)) 

for file in os.listdir(csv_folder): 
    if file.endswith('.csv'): 
        bus_number = file_to_bus.get(file.split('.')[0], None) 
        if bus_number is not None: 
            file_path = os.path.join(csv_folder, file)
            df = pd.read_csv(file_path, usecols=['itcs_stopName']) 
            for stop in df['itcs_stopName']: 
                stops_to_buses_counts[stop][bus_number] += 1 
                
stops_to_buses_min_10 = {stop: [bus for bus, count in buses.items() if count >= 10] 
                         for stop, buses in stops_to_buses_counts.items()} 

stops_to_buses_min_10_df = pd.DataFrame([(stop, ', '.join(map(str, buses))) for stop, buses in stops_to_buses_min_10.items()], 
                                        columns=['StopName', 'Buses']) 
```

### 9. For each (route, bus) pair, compute the ratio between the overall energy consumption and the overall driven distance

```python
df_in['busRoute'] = df_in['busRoute'].astype(str)
```

```python
route_bus_ratio = df_in.groupby(['busNumber', 'busRoute']).apply(lambda x: x['energyConsumption'].sum() /x['drivenDistance'].sum()).reset_index(name='energy_to_distance_ratio')
```

```python
route_bus_ratio
```

### 10. Starting from the results of the previous point, for each route compute the buses with max and min energy ratio, and save the difference between these ratios in a dataframe.

```python
energy_ratio_minmax = route_bus_ratio.groupby('busRoute').apply(
                                              lambda x: pd.DataFrame({
                                                  'busRoute': [x['busRoute'].iloc[0]],
                                                  'min_busNumber': [x.loc[x['energy_to_distance_ratio'].idxmin(), 'busNumber']],
                                                  'min_ratio': [x['energy_to_distance_ratio'].min()],
                                                  'max_busNumber': [x.loc[x['energy_to_distance_ratio'].idxmax(), 'busNumber']],
                                                  'max_ratio': [x['energy_to_distance_ratio'].max()],
                                                  'ratio_difference': [x['energy_to_distance_ratio'].max() - x['energy_to_distance_ratio'].min()]
                                              })).reset_index(drop=True)
```

```python
energy_ratio_minmax
```

### 11. Find the bus maximizing the difference computed in the previous point.

```python
max_diff_row = energy_ratio_minmax.loc[energy_ratio_minmax['ratio_difference'].idxmax()]
```

### 12. Extract the rows of the details such that the gnss_altitude differs from the value in the preceding row. Store also the difference in the variable altitude_variation.

```python
filtered_dataframes = []

for file in os.listdir(csv_folder):
    if file.endswith('.csv'):
        file_path = os.path.join(csv_folder, file)

        df = pd.read_csv(file_path, usecols=['gnss_altitude'])

        df['altitude_variation'] = df['gnss_altitude'].diff()

        filtered_df = df[df['altitude_variation'] != 0].copy()

        if not filtered_df.empty:
            filtered_df['file'] = file  # Aggiungo una colonna per il nome del file
            filtered_dataframes.append(filtered_df)

final_filtered_df = pd.concat(filtered_dataframes)
```

```python
final_filtered_df
```

### 15. For each bus compute the total time when the halt brake is active and the total time when the park brake is active. Compute also the ratio between those two times.


```python
brake_times = {} 

for file in os.listdir(csv_folder): 
    if file.endswith('.csv'): 
        bus_number = file_to_bus.get(file.split('.')[0], None) 
        if bus_number is not None: 
            file_path = os.path.join(csv_folder, file) 
            df = pd.read_csv(file_path, usecols=['time_unix', 'status_haltBrakeIsActive', 'status_parkBrakeIsActive']) 
            df = df.sort_values('time_unix') 
            df['time_diff'] = df['time_unix'].diff().fillna(0) 
            halt_brake_time = df.loc[df['status_haltBrakeIsActive'] == 1, 'time_diff'].sum() 
            park_brake_time = df.loc[df['status_parkBrakeIsActive'] == 1, 'time_diff'].sum() 
            brake_times[bus_number] = { 'halt_brake_time': halt_brake_time, 
                                       'park_brake_time': park_brake_time, 
                                       'ratio': halt_brake_time / park_brake_time if park_brake_time != 0 else None } 
brake_times_df = pd.DataFrame.from_dict(brake_times, orient='index').reset_index() 
brake_times_df.columns = ['bus_number', 'halt_brake_time', 'park_brake_time', 'ratio']
```

```python
brake_times_df
```

### 16. For each pair of stops that are consecutive in at least a trip, compute the average speed achieved when going from the first to the second stop.

```python
stop_pairs_speed = []

# Funzione per calcolare la velocità media tra due fermate consecutive
def calculate_average_speed(row):

    dist_1 = (row['gnss_latitude'], row['gnss_longitude'])
    dist_2 = (row['gnss_latitude_next'], row['gnss_longitude_next'])
    distance = geodesic(dist_1, dist_2).km


    time_diff = (row['time_unix_next'] - row['time_unix']) / 3600  # da secondi a ore
   

    if time_diff > 0:
        return distance / time_diff
    else:
        return None


for file in os.listdir(csv_folder):
    if file.endswith('.csv'):
        file_path = os.path.join(csv_folder, file)

        
        df = pd.read_csv(file_path, usecols=['itcs_stopName', 'time_unix', 'gnss_latitude', 'gnss_longitude'])

        
        df.dropna(subset=['itcs_stopName', 'time_unix', 'gnss_latitude', 'gnss_longitude'], inplace=True)

        
        df.sort_values('time_unix', inplace=True)

        # Creare le colonne shiftate per rappresentare le fermate successive
        df['itcs_stopName_next'] = df['itcs_stopName'].shift(-1)
        df['time_unix_next'] = df['time_unix'].shift(-1)
        df['gnss_latitude_next'] = df['gnss_latitude'].shift(-1)
        df['gnss_longitude_next'] = df['gnss_longitude'].shift(-1)

        
        df_filtered = df[df['itcs_stopName'] != df['itcs_stopName_next']].dropna()

        
        df_filtered['average_speed_kmh'] = df_filtered.apply(calculate_average_speed, axis=1)

        # Aggiungere il nome del file per riferimenti
        df_filtered['file'] = file


        stop_pairs_speed.append(df_filtered[['itcs_stopName', 'itcs_stopName_next', 'average_speed_kmh', 'file']])

final_df = pd.concat(stop_pairs_speed, ignore_index=True)
```

```python
final_df
```

```python
grouped_speed_df = final_df.groupby(['itcs_stopName', 'itcs_stopName_next'])['average_speed_kmh'].mean().reset_index()
```
